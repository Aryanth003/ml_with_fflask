```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import matplotlib.pyplot as plt
```

```python
data = {'age': [25, 30, 28, 35, 22],
        'gender': ["female", "male","male", "female","male"],
        'Status': ["Active",'InActive','Active','Active','InActive'],
        "rank":[1,3,4,2,5]}
l=LabelEncoder()
df=pd.DataFrame(data)
df['gender']=l.fit_transform(df['gender'])
df['Status']=l.fit_transform(df['Status'])
df['rank']=l.fit_transform(df['rank'])
df
```

```python
data = {'age': [25, 30, 28, 35, 22],
        'gender': ["female", "male","male", "female","male"],
        'Status': ["Active",'InActive','Active','Active','InActive'],
        "city":["hyderabad","delhi","banguluru","Ap","kerala"]}
d=pd.DataFrame(data)
m=OneHotEncoder(sparse_output=False)
col=["city"]
final = m.fit_transform(d[col])
m.get_feature_names_out(col)
mdf=pd.DataFrame(final,d['city'].values)
# cg=pd.DataFrame(final,d.col)
mdf
```
