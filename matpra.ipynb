```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=[5,7,7,8,9,10,10,10,11,12,12]
plt.hist(data,bins=len(set(data))+1,edgecolor='green')

plt.title('simple histogram example')
plt.xlabel('data')
plt.ylabel('frequency')
plt.show()
```

```python
data=[5,7,7,8,9,10,10,10,11,12,12]
plt.plot(data)

plt.title('simple histogram example')
plt.xlabel('data')
plt.ylabel('frequency')
plt.show()
```

```python
data=[5,7,7,8,9,10,10,10,11,12,12]
for i in data:
    plt.plot(data.index(i),i,'o')

plt.title('simple histogram example')
plt.xlabel('data')
plt.ylabel('frequency')
plt.show()
# plt.hist(data,bins=len(set(data))+1,edgecolor='green',color='yellow')

# plt.title('simple histogram example')
# plt.xlabel('data')
# plt.ylabel('frequency')
# plt.show()
```
