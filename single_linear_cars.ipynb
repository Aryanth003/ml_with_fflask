```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Cars.csv")
```

```python

correlation = df['HP'].corr(df['SP'])
print(f"The correlation between hp and mpg is: {correlation}\n\n\n")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='HP', y='SP', data=df)
plt.title('hp vs SP')
plt.xlabel('Horsepower')
plt.ylabel('SP')
plt.show()
```

```python
import statsmodels.formula.api as sm
model=sm.ols(formula="SP~HP",data=df).fit()
```

```python
print(model.summary())
```
